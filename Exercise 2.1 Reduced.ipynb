```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage

# Step 2: Load the unscaled data
path = '/Users/andrewfearney27/Downloads/Dataset-weather-prediction-dataset-processed.csv'
weather_data = pd.read_csv(path)

# Step 3: Reduce the dataset by selecting fewer columns and filtering specific years
# Adjust the columns and rows selection based on your analysis
reduced_data = weather_data.drop(columns=['DATE', 'MONTH'], errors='ignore')  # Drop unnecessary columns
reduced_data = reduced_data.sample(100)  # Taking a sample of 100 rows for demonstration (you can adjust this)

# Step 4: Perform scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(reduced_data)
```

```python
# Step 5: Apply PCA to reduce the dimensionality to a manageable level (e.g., 9 components)
pca = PCA(n_components=9)
pca_data = pca.fit_transform(scaled_data)

# Step 6: Create a DataFrame with PCA components
pca_df = pd.DataFrame(pca_data, columns=[f'PCA{i+1}' for i in range(9)])
```

```python
# Step 7: Create dendrograms using different linkage methods

# Dendrogram with 'single' linkage
distance_single = linkage(pca_df, method='single')
plt.figure(figsize=(18, 6))
dendrogram(distance_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.title('Dendrogram Single Method', fontsize=18)
plt.show()
```

```python
# Dendrogram with 'complete' linkage
distance_complete = linkage(pca_df, method='complete')
plt.figure(figsize=(18, 6))
dendrogram(distance_complete, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.title('Dendrogram Complete Method', fontsize=18)
plt.show()
```

```python
# Dendrogram with 'average' linkage
distance_average = linkage(pca_df, method='average')
plt.figure(figsize=(18, 6))
dendrogram(distance_average, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.title('Dendrogram Average Method', fontsize=18)
plt.show()
```

```python
# Dendrogram with 'ward' linkage
distance_ward = linkage(pca_df, method='ward')
plt.figure(figsize=(18, 6))
dendrogram(distance_ward, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.title('Dendrogram Ward Method', fontsize=18)
plt.show()
```
